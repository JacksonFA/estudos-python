# Projeto Ciência de Dados - Previsão de Vendas

- Nosso desafio é conseguir prever as vendas que vamos ter em determinado período com base nos gastos em anúncios nas 3 grandes redes que a empresa Hashtag investe: TV, Jornal e Rádio
- TV, Jornal e Rádio estão em milhares de reais
- Vendas estão em milhões

#### Importar a Base de dados

```python
import pandas as pd

tabela = pd.read_csv('advertising.csv')

display(tabela)
```

#### Análise Exploratória
- Vamos tentar visualizar como as informações de cada item estão distribuídas
- Vamos ver a correlação entre cada um dos itens

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o gráfico
sns.heatmap(tabela.corr(), cmap='Wistia', annot=True)

# Exibindo o gráfico
plt.show()
```

#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
# Quem vai ser usado para a previsão
x = tabela[['TV', 'Radio', 'Jornal']]

# Quem vai ser previsto
y = tabela['Vendas']

from sklearn.model_selection import train_test_split as ttspt

x_treino, x_teste, y_treino, y_teste = ttspt(x, y, test_size=0.3)
```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression as linear
from sklearn.ensemble import RandomForestRegressor as forest

# Cria a IA
modelo_regressao = linear()
modelo_arvore = forest()

# Treina a IA
modelo_regressao.fit(x_treino, y_treino)
modelo_arvore.fit(x_treino, y_treino)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
# Previsão nos testes
previsao_regressao = modelo_regressao.predict(x_teste)
previsao_arvore = modelo_arvore.predict(x_teste)

# Teste das previsões
from sklearn.metrics import r2_score as r2
print(r2(y_teste, previsao_regressao))
print(r2(y_teste, previsao_arvore))
```

#### Visualização Gráfica das Previsões

```python
tabela_aux = pd.DataFrame()
tabela_aux['y_teste'] = y_teste
tabela_aux['arvore decisao'] = previsao_arvore
tabela_aux['regressao linear'] = previsao_regressao

plt.figure(figsize=(15,6))
sns.lineplot(data=tabela_aux)
plt.show()
```
